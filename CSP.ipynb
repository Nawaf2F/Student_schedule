# 1. Introduction
---
- Constraint satisfaction problems are mathematical questions. they are defined as set of objects which their state must satisfy certain constraints or limits.
- constraint satisfaction methods solve homogeneous collection of finite constraints over variables that are representation of the problem entities.
- constraint satisfaction problems often exhibit high complexity, to solve it in reasonable time a combination of heuristics and the combinatorial search is required.
- All constrain satisfaction problems are commutative that means the order of assigned variables doesn’t matter.
- This tutorial is to explain constraint satisfaction problems solving methods and how implement them to solve the problems.

- We decided that we're gonna work on the **student problem** assuming that the university uploaded the sections that you are gonna be able to register for this semester, with each section having a **code, subject name, section number, time.**


- What we're going to do is let the student enter the names of the subjects that he wants to register. After that we’re going to let the algorithm find a solution so that **there is no conflict between lecture times**, a solution found if there is a **complete assignment** .





- for simplicity we’re going to represent the lectures times form 1 to 25 were:
 -  1 to 5 are Sunday lectures.
 -  6 to 10 are Monday lectures.
 -  11 to 15 are Tuesday lectures.
 -  16 to 20 are Wednesday lectures.
 -  21 to 25 areThursday lectures.
 
---


---
           
- **initial state:** all of the variables is not assigned
- **action:** assign value to one of the variable
- **state space:** 
 - the variables
 - the domains
 - the constraints
- **Goal state:** all of the variables is assigned 
---
**variables:** is the subject that the student enters.
 { subject 1,subject 2,......,subject n }.

**domains:** each subject has a different domain where each value in the domain represented as a list => {1,2,11,12} this value represent:
- Sunday   lecture 1 and 2
- Tuesday  lecture 1 and 2


For example the domain of CS361 is { {3,4,13,14}  , {1,2,11,12} }.

**constraints:** there is no duplicate index every index in a value list is unique
and it doesn't appear in other values lists. For example we assigned values to two different subjects.


- CS361 = {1,2,21,**22**}, CS390 = {11,12,**22**,23}  **22** is duplicate so the constraint is not satisfied.



- CS361 = {1,2,**21**,**22**}, CS390 = {11,12,**21**,**22**}  **21** and **22** are duplicates so the constraint is not satisfied.


- CS361 = {1,2,13,14}, CS390 = {11,12,21,22}  no duplicate values so this assignment satisfies the constraint.

```python
class section:

    def __init__(self, code, subject_name, section_number, time):

        self.code = code
        self.subject_name = subject_name
        self.section_number = section_number
        self.time = time

class Variable:

    def __init__(self, name, domain_values):
        self.name = name
        self.domain_values = domain_values #list of available sections        
```

```python
class Schedule_registration:

    def __init__(self, Available_titles, student_requests):

        self.Available_titles = Available_titles #this list represent the sections available for registration
        self.student_requests = student_requests #this list represent the subjects names
        self.listOf_Variables = []

    def set_up_Variables_domains(self, null=None):

        for name in self.student_requests:
            self.listOf_Variables.append(Variable(name , null))

        for i, v in enumerate(self.listOf_Variables):
            section_list = []
            for sec in self.Available_titles:
                if (str(v.name) == str(sec.subject_name)):
                    section_list.append(sec)

            self.listOf_Variables[i].domain_values = list(section_list)
            section_list.clear()

    def check_constraint(self, assigment_list) -> bool :

        count = 0
        all_times = set()
        for key, value in assigment_list.items():
            if value != None:
                count += len(value.time)
                all_times = all_times.union(value.time)
                if (len(all_times) != count):
                    return False

        return True
```

# 3.1 Backtracking Explanation
---
- Uses depth first search to choose values for variables at a time and backtracks when variables have no legal values left to assign. it repeatedly chooses an unassigned variable and then tries all values in the domain of that value in turn trying to find solution if inconsistency is detected it returns failure causing it to go back and try another value it doesn't need to be supplied with specific initial state ,action function or goal test. Time complexity of the Standard backtrack algorithm is too high
---

# Backtrack search pseudocode description 
---
### 1. The first function "BACKTRACKING-SEARCH":

- takes a **csp** in the parameter
- calls on another function **"RECURSIVE-BACKTRACKING"** sending an **empty list and the csp**
- in the end it will **return "Solution"** if solution is found or **return "Failure"** if there is no solution

### 2. The second function "RECURSIVE-BACKTRACKING":

- takes an empty list(**"assignment"**) and a csp in as a parameters that is coming from the **"BACKTRACKING-SEARCH"** function
- the **base case** of the recursive function is when all of the variables have been assigned
- we put the list of unassign variables into a local variable(**"var"**) to assign the values to it
- then we check if the **"value"** is consistant with the **"assignment"** given the constraints
- if the **"value"** is inconsistant with the **"assignment"** then **return failure**
- if not, assign the **"value"** to **"var"** and add it to the **"assignment list"**
- the function call itself and this is the **recursive case** and we put the result into a local variable(**"result"**)
- if a solution is found **return the "result"** and **clear the assignment list**

### 2. الوظيفة الثانية " RECURSIVE-BACKTRACKING ":

- تأخذ لستة فارغة (**"لستة التعيينات"**) و **مشكلة القيود**  كمعطى قادم من وظيفة **"BACKTRACKING-SEARCH"**

- نضع لستة المتغيرات التي لم نسند لها قيم في متغير محلي (**"var"**) لإسناد القيم إليه

- إذا لم يكن الأمر كذلك ، فقم بتعيين **"القيمة"** إلى **"var"**وأضفها إلى**"لستة التعيينات"**

```python
    def backtracking_search( rqst, assignment = {}):
        
        if len(assignment) == len(rqst.listOf_Variables):
            return assignment

        unassigned= [v for v in rqst.listOf_Variables if v not in assignment]
        
        first = unassigned[0]
        for value in first.domain_values:
            local_assignment = assignment.copy()
            local_assignment[first] = value
            
            if rqst.check_constraint(local_assignment):
                result= backtracking_search(rqst,local_assignment)
                
                if result is not None:
                    return result
        return None
```

# Backtracking Code Description
---
### The code above is for the backtracking search
- The function **"backtracking_search"** is a recursive function that takes an **"assignment list"**
- the first line is the **base case** for the recursive function where we compare the length of **"assignment list"** and the length of the **"requist.listOf_Variables"** if it is equal then **return the assignment**
- we store the unassigned variables in the **"unassigned list"**
- initiate a local variable **"first"** and store in it **"unassigned[0]"** 
- a loop that run through the **"first.domain_values"**
- initiate another local variable **"local_assignment"** that takes a copy from **"assignment"** list
- we assign the variables that has not been assigned by assigning a value to **"local_assignment[first]"**
- then we check the constraint by using **"requist.check_constraint(local_assignment)"**
- if the **"local_assignment"** does not satisify the constraint then we do not have a solution **return none**
- then the **recursive case**, the function will call itself and store the result in variable **"result"**
- if there is a solution **return "result"**

# 5.1 AC-3 Explanation
---
arc:
A variable in CSP is arc consistent if all domain satisfies variables constraints. As for every value of X there's some allowed in Y. if X loses a value, X neighbors must be rechecked. Arc consistency aims to detect failure early. can be run as a preprocessor or after each assignment or as a preprocessor before each start. it must run rapidly until no inconsistency remains. arc checking can be done in **O(d^2)** and the worst-case time complexity is **O(cd^3)**.

---

# AC-3 Pseudocode Description
---
- X = xi ,y = xj
1. The AC-3 function accepts CSP as an argument and contains variables, domains, and constraints.

2. We will create a local variable queue that will contain all of the arcs.The arcs are a list of all the constraints from both sides; for example, if A > b, the queue will have the constraint from both directions as fallow.(A > B, B < A)

3. Iterate through the queue and call the function Revise for each arc in the queue.If the function Revise returned True, we will check if the domain of x is empty; if it is, there is no arc consistency, and the function AC 3 will return false; otherwise, we will add all the arcs pointing to x to the queue.

4. If the AC-3 returned true, this indicates that the CSP is arc consistent.

5. The Revise function takes an arc x and y and loops through the x domain values; if there is a value in the x domain that does not have a value in the y domain that satisfies the constraints between them, the function Revise returns True; otherwise, it returns false.

# 5.2 AC-3 Implementation
---
# <div style="direction:rtl">  5.2 تنفيذ كود خوارزمية تطابق الاقواس-3 </div>
---

```python
def check_arc(x,y)-> bool :

    return len(x.union(y)) == len(x) + len(y)

def revise(x, y):
   
    revised = False

    x_domain = x.domain_values
    y_domain = y.domain_values

    for x_value in x_domain:
        satisfies = False
        for y_value in y_domain:
            if check_arc(x_value.time, y_value.time):
                satisfies = True
        if not satisfies:
            x_domain.remove(x_value)
            revised = True

    return revised


def ac3(arcs):

    queue = arcs[:]

    while queue:
        (x, y) = queue.pop(0)
      
        revised = revise(x, y)
      
        if revised:
            
            if len(x.domain_values) == 0:
                return False
           
            for neighbor in arcs:
                if neighbor[1] == x:
                    queue.append(neighbor)
    return True


def create_arcs(list_of_Variables):

    arcss = []
    for x in list_of_Variables:
        for y in list_of_Variables:
            if(x != y):
                arcss.append((x , y))

    return arcss 
```

# 6. Forward Checking
---
# <div style="direction:rtl"> 6. خوارزمية التحقق الامامي</div>

---
# 6.1 Forward Checking Explaination
---
Forward checking is the easiest way to prevent conflicts. it performs restricted form of arc consistency to the not yet instantiated variables, we used restricted arc consistency because forward checking checks only the constraints between current and future variables. When a value is assigned to a variable, any value in the domain of the future variables which conflict with this assignment is removed from the domain temporarily. it uses this so that the domain of the future variables becomes empty. then it is known that current solution is inconsistent. it allows leading to failure branches of search tree to be pruned earlier than simple backtracking, then the search tree and the overall work is reduced. when a new variable is assigned, old remaining variables must be consistent with past variables, so checking after each assignment is no longer necessary. Forward checking is much better choice than simple backtracking.

# <div style="direction:rtl"> 6.1 خوارزمية التحقق الامامي</div>
---
<div style="direction:rtl">
التحقق الأمامي هو أسهل طريقة لمنع التعارض. إنه يؤدي شكلا مقيدا من تناسق القوس إلى المتغيرات التي لم يتم إنشاؤها بعد، واستخدمنا تناسق القوس المقيد لأن الفحص الأمامي يتحقق فقط من القيود بين المتغيرات الحالية والمستقبلية. عندما يتم تعيين قيمة إلى متغير، تتم إزالة أي قيمة في مجال المتغيرات المستقبلية التي تتعارض مع هذا التعيين من المجال مؤقتا. يستخدم هذا بحيث يصبح مجال المتغيرات المستقبلية فارغا. ثم من المعروف أن الحل الحالي غير متناسق. يسمح بتقليم الفروع المؤدية إلى الفشل خارج شجرة البحث في وقت أبكر من التراجع البسيط., ثم يتم تقليل شجرة البحث والعمل الكلي. عند تعيين متغير جديد، يجب أن تكون المتغيرات القديمة المتبقية متناسقة مع المتغيرات السابقة، لذلك لم يعد التحقق بعد كل تعيين ضروريا. التحقق الأمامي هو خيار أفضل بكثير من التراجع البسيط.
